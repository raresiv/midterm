### Variable Explanation (Do not change variable names)
- 'w' is true coefficients of the original polynomial function
- 'original_degree' is the order of the original polynomial function
- 'x_list' is a list of the points at $x$-axis

### Our goal is to estimate 'w' from data points, 'x_data' and 'y_data'. Answer the following problems.

### 1. Compute a Vandermonde matrix when the degree of polynomial is $4$ (30pt)
- The variable 'degree' is the order of polynomial. In this problem, we set degree=$4$
- Use the variable 'A' for the Vandermonde matrix. Now, 'A' is initialized as a zero matrix whose elements are all zero. Fill in the element of the Vandermonde matrix by using power operator (\*\*), for loop, and np.concatenation.

```python
degree = 4
A = np.zeros((n_data, degree+1)) # Dummy initialization
for i in range(n_data):
    for j in range(degree + 1):
        A[i, j] = x_data[i] ** j
#A=np.round(A,decimals=1)
```

### Print results (do not modify the following block)

```python
print(A)
```

### 2. Compute the coefficients of polynomial regression using a $4$ degree polynomial  (40pt)
- Use the variable 'degree' and the Vandermonde matrix 'A' in Problem 1.
- The variable 'w_est' is the coefficients of polynomial regression. Now, 'w_est' is initialized as a zero vector. Compute the 'w_est' from 'A' and 'y'
- The variable 'y_est' is an estimated function value corresponding to the input points 'x_list'. Now, it is a zero list and fill the list by computing the estimated function values. In other words, y_est = $\hat{f}($x_list$)$

```python
w_est = np.zeros((degree+1,1))
y_est = np.zeros_like(x_list)

# Compute the SVD of A
U, S, Vt = np.linalg.svd(A, full_matrices=False)

# Pseudo-inverse of S
S_inv = np.diag(1.0 / S)

# Compute w_est using SVD
w_est = Vt.T @ S_inv @ U.T @ y_data

for i in range(len(x_list)):
    for j in range(degree + 1):
        y_est[i] += w_est[j] * (x_list[i] ** j)
```

### Print results (do not modify the following block)

```python
plt.plot(x_list, y_list, 'b:', linewidth=2, label="Original Function")
plt.plot(x_list, y_est, 'm-', linewidth=2, label="Polynomial Regression (d={})".format(degree))
plt.scatter(x_data, y_data, s=50, c='r', label="Data Points")
plt.xlim([np.min(x_list),np.max(x_list)])
plt.ylim([np.min(y_data),np.max(y_data)])
plt.legend(prop={'size': 12})
plt.title("Data Plot")
plt.show()
```

Using a high-degree polynomial in a polynomial regression can closely fit complex and intricate patterns in the data. They are extremely flexible and can approximate a wide range of functions. They also have low bias, which means they can closely match the training data, reducing the error on the training set. However, high-degree polynomials are prone to overfitting. They can capture noise in the data and generalize poorly to unseen data. Overfit models tend to have high variance, leading to large variations in predictions for different training sets.

Low-degree polynomials are simpler and have a smoother, more interpretable shape. They are also less likely to overfit, making them better at generalizing to new, unseen data. Low-degree polynomials typically have lower variance, leading to more stable and consistent predictions. However, Low-degree polynomials may underfit the data, meaning they cannot capture the underlying patterns in the data accurately, resulting in a high training error. Furthermore, they also might not be able to represent complex relationships in the data.

This phenomen is called bias-variance trade-off in machine learning. The bias-variance trade-off in machine learning refers to the need to balance the bias (simplification) and variance (flexibility) of a model. High-degree polynomials exhibit high variance and low bias, while low-degree polynomials have low variance and high bias. The goal is to find a degree that strikes a balance between fitting the data well and generalizing to new, unseen data.
